```python
import pandas as pd
import numpy as np
```

```python
# Create a simple dataset
data = {
    'A': [10,20,3,40,50,60,70,80,90,100],
    'B': [0.1,0.2,0.3,0.4,0.50,0.6,0.7,0.8,0.9,0.01]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Standarize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
# Apply pca
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit_transform(scaled_data)
```

```python
# Create a Dataframe using Principle Component
pca_df = pd.DataFrame(pca.fit_transform(scaled_data), columns = ['PC1','PC2'])
pca_df
```

```python
# Check explained variance
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance = {explained_variance}")
```

```python
# Visualize the Principle Components
import matplotlib.pyplot as plt
plt.scatter(pca_df['PC1'],pca_df['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

# Difference between fit & fit_transform

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Sample Data
x = np.array([[1, 2], [3, 4], [5, 6]])
print(x)
```

```python
# using fit
scaler = StandardScaler()
scaler.fit(x)
print('Mean: ', scaler.mean_)
print('variance: ', scaler.var_)
print('scale: ', scaler.scale_)
```

```python
# Now we are transforming the data using transform alone
x_scaled = scaler.transform(x)
print(x_scaled)
```

```python
# using fit_transform
x_scaled = scaler.fit_transform(x)
print(f"Transformed using fit_transform method : {x_scaled}")
```
